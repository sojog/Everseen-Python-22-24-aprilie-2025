```python

class MyClass:

    def __init__(self, price):
        self.__price = price

    def get_price(self):
        return self.__price

    def set_price(self, new_value):
        self.__price = new_value
```

```python
obj = MyClass(100)
print(obj.get_price())
obj.set_price(10)
print(obj.get_price())
```

```python

class MyClass:

    def __init__(self, price):
        self.__price = price

    @property
    def get_price(self):
        return self.__price

    def set_price(self, new_value):
        self.__price = new_value
```

```python
obj = MyClass(100)
print(obj.get_price)
obj.set_price(10)
print(obj.get_price)
```

```python

class MyClass:

    def __init__(self, price):
        self.__price = price

    @property
    def price(self):
        return self.__price

    def set_price(self, new_value):
        self.__price = new_value


obj = MyClass(100)
print(obj.price)
obj.set_price(10)
print(obj.price)
```

```python

class MyClass:

    def __init__(self, price):
        self.__price = price

    @property
    def price(self):
        return self.__price
    
    @price.setter
    def price(self, new_value):
        if isinstance(new_value, int):
            self.__price = new_value


obj = MyClass(100)
print(obj.price)
obj.price = 24125235
print(obj.price)
obj.price = "whatever"
print(obj.price)
```
