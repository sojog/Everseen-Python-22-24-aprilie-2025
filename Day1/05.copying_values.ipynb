# Deep copy

```python
list1 = [1, 2, 3]
list2 = [1, 2, 3]

list1 == list2
```

```python
id(list1) == id(list2)
```

```python
list1 is list2
```

```python
list3 = list1[:]
id(list3) == id(list1)
```

```python
list1 = [1, 2, 3]
list2 = [1, 2, 3]


list_comp = [list1, list2]
list_comp
```

```python
list_copy = list_comp[:]
list_copy
```

```python
list_copy[0].append(10)
```

```python
list_copy
```

```python
list_comp
```

```python
import copy
```

```python
list_deep = copy.deepcopy(list_comp)
list_deep
```

```python
list_deep[0].append(100)
```
