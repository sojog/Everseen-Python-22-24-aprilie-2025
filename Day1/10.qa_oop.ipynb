```python
class MyClass:
    def __init__(self):
        # public var -> Inside & Subclasses & Outside    
        self.x = 10
        # internal/procted var -> Inside & Subclasses
        self._y = 20
        # private var -> Inside
        self.__z = 30
```

```python
my_obj = MyClass()
print("x=", my_obj.x)
```

```python
print("y=", my_obj._y)
```

```python
print("z=", my_obj.__z)
```

```python
print("z=", my_obj._MyClass__z)
```

```python
class MyClass:
    def __init__(self):
        # public var -> Inside & Subclasses & Outside    
        self.x = 10
        # internal/procted var -> Inside & Subclasses
        self._y = 20
        # private var -> Inside
        self.__z = 30


class SubClass(MyClass):

    def print(self):
        print(self.__z)
```

```python
obj = SubClass()
obj.print()
```

```python
obj._SubClass__z
```

```python
obj._MyClass__z
```
