```python
class A:
    __slots__ = ('__x', '__y')

    def __init__(self, x=5, y=6):
        self.__x  = x
        self.__y = y

    def __str__(self):
        return f"{self.__x} {self.__y}"

    @property
    def x(self):
        return self.__x
    
    @x.setter
    def x(self, new_value):
        self.__x = new_value

obj_A = A()
obj_A._A__x = "hello"
obj_A._A__y = 123


obj_A.x = [231, 32, 320]

print(obj_A)
```
