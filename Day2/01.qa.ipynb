```python
def outer(x):
    def inner():
        nonlocal x
        x += 1
        return x
    return inner

f = outer(10)
print(f())
print(f())
```

```python
class A:
    def __init__(self, x):
        self.x = x
    def __mul__(self, other):
        return A(self.x * other.x)

a = A(3)
b = A(4)
print((a * b).x)
```

```python
class Animal:
    def speak(self):
        return "..."

class Dog(Animal):
    def speak(self):
        return "Woof"

class Cat(Animal):
    def speak(self):
        return "Meow"

animals = [Dog(), Cat(), Animal()]
print(" ".join(a.speak() for a in animals))
```

```python
class A:
    def __init__(self, x):
        self.x = x
    def __eq__(self, other):
        return self.x == other.x
    def __hash__(self):
        return hash(self.x)

s = {A(1), A(1), A(2)}
print(len(s))
```

```python
class A:
    def __init__(self, x):
        self.x = x
    def __hash__(self):
        return hash(self.x)

a1 = A(1)
a2 = A(1)
a3 = A(3)
a4 = a3
print("ids:", id(a1), id(a2), id(a3))
print("hash:", hash(a1), hash(a2), hash(a3))
s = {a1, a2, a3, a4}
print(len(s))
```

```python
class A:
    def __init__(self, x):
        self._x = x

    @property
    def x(self):
        return self._x * 2

a = A(5)
print(a.x)
```

```python
class Mixin:
    def hello(self):
        return "Hello from Mixin"

class Base:
    def hello(self):
        return "Hello from Base"

class Child(Mixin, Base):
    pass

print(Child().hello())
```

```python
class A:
    def __iadd__(self, other):
        self.x += other
        return self

a = A()
# a.x = 5
a += 3
print(a.x)
```

```python
class MyClass:
    def __init__(self):
        self.x = 10

    def create_an_attr_later(self):
        self.y = 100


a = MyClass()
a.create_an_attr_later()
print("Y from a", a.y)

b = MyClass()
print("Y from b", b.y)
```
