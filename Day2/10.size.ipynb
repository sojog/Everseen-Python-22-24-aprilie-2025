```python
class PersonWithSlots:
    __slots__ = ['name', 'age']
    def __init__(self, name, age):
        self.name = name
        self.age = age


class PersonWithoutSlots:
    def __init__(self, name, age):
        self.name = name
        self.age = age

object_with  = PersonWithSlots("Maria", 20)
object_without  = PersonWithoutSlots("Maria", 20)
print("With Slots:", sys.getsizeof(object_with))
print("Without Slots:", sys.getsizeof(object_without))
```

```python
class PersonWithSlots:
    __slots__ = ['name', 'age', "address"]
    def __init__(self, name, age, address):
        self.name = name
        self.age = age
        self.address = address


class PersonWithoutSlots:
    def __init__(self, name, age, address):
        self.name = name
        self.age = age
        self.address = address

        
```

```python
object_with  = PersonWithSlots("Maria", 20, "Timisoara")
object_without  = PersonWithoutSlots("Maria", 20, "Timisoara")
```

```python
print("With Slots:", sys.getsizeof(object_with))
print("Without Slots:", sys.getsizeof(object_without))
```

```python
class PersonWithSlots:
    __slots__ = ['name', 'age', "address","another"]
    def __init__(self, name, age, address):
        self.name = name
        self.age = age
        self.address = address
        self.another = "hello"


class PersonWithoutSlots:
    def __init__(self, name, age, address):
        self.name = name
        self.age = age
        self.address = address
        self.another = "hello"


object_with  = PersonWithSlots("Maria", 20, "Timisoara")
object_without  = PersonWithoutSlots("Maria", 20, "Timisoara")
```

```python
print("With Slots:", sys.getsizeof(object_with))
print("Without Slots:", sys.getsizeof(object_without))
```

```python
print("With Slots:", sys.getsizeof(object_with))
print("Without Slots:", sys.getsizeof(object_without) + sys.getsizeof(object_without.__dict__))
```
