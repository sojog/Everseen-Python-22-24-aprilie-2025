```python
class Person:
    __slots__ = ['name', 'age', "address"]
    __dict__ = {}
    def __init__(self, name, age, address):
        self.name = name
        self.age = age
        self.address = address

p1 = Person('Maria', 25, "Timisoara")
print(p1)
```
