```python
nums = [1,2,3,2]

freq = {}
for i in nums:
    freq[i] = freq.get(i, 0) + 1

total = 0
for i in freq:
    if freq[i] == 1:
        total += i
total

```

```python
def pythonic_sum(nums):
    freq = {}
    for i in nums:
        freq[i] = freq.get(i, 0) + 1

    total = 0
    for i in freq:
        if freq[i] == 1:
            total += i
    return total
```

```python
pythonic_sum([1,2,3,2])
```

```python
from time import perf_counter
```

```python
start = perf_counter()
pythonic_sum([1,2,3,2])
stop = perf_counter()

stop - start
```

```python
from collections import Counter
```

```python
nums = [1,2,3,2]
Counter(nums)
```

```python
def counter_func(nums):
    counts = Counter(nums)
    return sum([i for i  in counts if counts[i]== 1])
```

```python
counter_func(nums)
```

```python
pythonic_start = perf_counter()
pythonic_sum(nums)
pythonic_stop = perf_counter()
```

```python
counter_start = perf_counter()
counter_func(nums)
counter_stop = perf_counter()
```

```python
pythonic_stop - pythonic_start < counter_stop - counter_start
```
