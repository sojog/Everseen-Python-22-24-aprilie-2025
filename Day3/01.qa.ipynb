```python
import bisect

a = [1, 3, 5, 7, 9]
bisect.insort(a, 4)
print(a)
```

```python
from functools import reduce
from operator import mul

result = reduce(mul, [1, 2, 3, 4, 5])
print(result)
```

```python
from collections import deque

q = deque([1, 2, 3])
q.appendleft(0)
print(list(q))
```

```python

from itertools import islice

def naturals():
    n = 0
    while True:
        yield n
        n += 1

print(list(islice(naturals(), 5)))
```

```python
from typing import Self


def add(x: int | float, y: int | float) -> int:
    return x + y


add("2", "3")
```

```python
class Shape:
    def __init__(self):
        self.scale = 0.0
    def set_scale(self, scale: float) -> Self:
        self.scale = scale
        return self
    

t = Shape()
type_of = type(t.set_scale(30))
print(type_of)
```

```python
def deco1(cls):
    cls.val = 1; return cls

def deco2(cls):
    cls.val = 2; return cls


@deco1
@deco2
class A:pass
print(A.val)
```
