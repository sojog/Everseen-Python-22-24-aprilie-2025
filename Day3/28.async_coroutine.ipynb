```python
import inspect
```

```python
async def f(): # declaration of a coroutine
    return 123
```

```python
type(f) # <class 'function'>
```

```python
inspect.isbuiltin(print)
```

```python
x = 10
inspect.isfunction(x)
```

```python
import numpy as np
print(inspect.getdoc(np.ones))
```

```python
inspect.iscoroutinefunction(f)
```

```python
# identical to the way generator functions work in Python
def g():
    yield 123
```

```python
type(g) # <class 'function'>
```

```python
gen = g()
type(gen) # <class 'generator'>
```
